## Aggregate

```python
df.agg(["min","max","mean","std"])
```

```python
df.agg(["min","max","mean","std"])[["mpg","cyl","gear"]]
```

```python
df.agg({
    "hp" : ["max","min"],
    "cyl" : ["mean"],
    "gear" : ["mean","min","max"]
})
```

```python
df.groupby("cyl").agg({
    "hp" : ["max","min"],
    "gear" : ["mean","min","max"]
})
```
