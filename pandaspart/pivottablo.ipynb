# **Pivot Tablo**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("pokemon.csv")
```

```python
df.describe()
```

```python
pd.pivot(data=df, index="Name", columns="Legendary",values="Generation")
```

Pokemon türlerinin nesillere göre ortalama saldırı değeri nedir?

```python
pd.pivot_table(data=df, index="Type 1", columns="Generation",values="Attack")
```

Türlerin nesillere göre saldırı güçleri

```python
pd.pivot_table(data=df, index="Generation", columns="Type 1", values="Attack")
```

```python
df[(df["Type 1"] == "Dragon") & (df["Generation"] == 1)].mean(numeric_only=True)
```

```python
pd.pivot_table(data=df,index=["Generation","Legendary"],columns="Type 1",values="Attack")
```

```python
pd.pivot_table(data=df, index="Generation",aggfunc="sum")
```

```python
pd.pivot_table(data=df, index="Generation", aggfunc="min")
```

```python
pd.pivot_table(df, index="Type 1",values=["Attack","Defense","Sp. Atk","Sp. Def"],aggfunc="min")
```

```python
pd.pivot_table(df, index="Type 1",values =["Generation","Speed"])
```

```python
df.groupby("Type 1").mean(numeric_only="True")[["Generation","Speed"]]
```

```python
pd.pivot_table(df,index="Generation",columns="Type 1",
               values="Attack", fill_value=0)
```
