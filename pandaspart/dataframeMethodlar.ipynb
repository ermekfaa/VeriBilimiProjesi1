```python
def OP(power):
    if power < 320:
        return "*"
    elif power < 430:
        return "**"
    elif power < 500:
        return "***"
    else:
        return "*****"
```

```python
df["Star"] = df["Total"].apply(OP)
```

```python
def defense_power(Def, Sp_def):
    if Def + Sp_def > 240:
        return "Shield"
    else:
        return "Neutral"
```

```python
df["Defender"] = df[["Defense","Sp. Def"]].apply(lambda df:defense_power(df["Defense"], df["Sp. Def"]), axis=1)
```

```python
import numpy as np
```

```python
def attack_power(Atk, Sp_atk):
    if Atk + Sp_atk > 240:
        return "Sword"
    else:
        return "Neutral"
```

```python
df["Attacker"] = np.vectorize(attack_power)(df["Attack"],df["Sp. Atk"])
# Vectorize'ın içine fonksiyonu yanına değerleri
```

```python
df.iloc[427:437]
```

```python
df[df["Attacker"] == "Sword"]
```

```python
import timeit 
```

```python
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv("pokemon.csv")

def defense_power(Def, Sp_def):
    if Def + Sp_def > 240:
        return "Shield"
    else:
        return "Neutral"

'''

stmt_one = '''
df["Defender"] = df[["Defense","Sp. Def"]].apply(lambda df:defense_power(df["Defense"], df["Sp. Def"]), axis=1)
'''

stmt_two = '''
df["Defender"] = np.vectorize(defense_power)(df["Defense"],df["Sp. Def"])
'''


```

```python
timeit.timeit(setup=setup,
              stmt=stmt_one,
              number=1000)
```

```python
timeit.timeit(setup=setup,
              stmt=stmt_two,
              number=1000)
```
