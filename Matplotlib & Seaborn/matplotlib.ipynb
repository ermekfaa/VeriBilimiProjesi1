## Grafik d√ºzenleme

```python
x = np.arange(0,10)
y = x**2 - 10*x + 25
```

```python
x,y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize = (12,6),dpi=200)
plt.plot(x,y)
plt.title("Ben Ba≈ülƒ±k istiyorum",size=18)
plt.xlabel("X ekseniyim", size=20)
plt.ylabel("Y ekseniyim", size=20)
plt.xticks(np.arange(min(x),max(x) + 1,1));
plt.yticks(np.arange(min(y),max(y) + 1,2.5));
plt.show() # Ya da ; koy
```

dpi kaliteyle ilgili

## Fig√ºr ;Komutlarƒ±

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])

axes1.plot(x,y)
axes1.set_xlabel("x ekseni")
axes1.set_ylabel("y ekseni")
axes1.set_title("B√ºy√ºk Ba≈ülƒ±k")

axes2.plot(x,y)
axes2.set_xlabel("x ekseni")
axes2.set_ylabel("y ekseni")
axes2.set_title("K√º√ß√ºk Ba≈ülƒ±k")
```

```python
fig = plt.figure()

axes1= fig.add_axes([0,0,1,1])
axes2= fig.add_axes([0.7,0.7,0.2,0.2])

axes1.plot(x,y)
axes1.set_xlabel("x ekseni")
axes1.set_ylabel("y ekseni")
axes1.set_title("Fig√ºr")

axes2.plot(x,y)
axes2.set_ylim(0,5)
axes2.set_xlim(3,7)
axes2.set_title("Se√ßilen Alan")


fig.savefig("fig_kayit.png",bbox_inches="tight")
```

## Legend

```python
x = np.arange(0,10)
```

```python
plt.figure(figsize=(10,8), dpi=140)
plt.plot(x,x*2 + x, label = "2x+x")
plt.plot(x, x**2 - x, label = "x^2 - x")
plt.legend();
```

loc=

0 - en iyi yeri se√ßer

1 - saƒü √ºst

2 - sol √ºst

3 - sol alt

4 - saƒü alt

```python
plt.figure(figsize=(10,8), dpi=140)
plt.plot(x,x*2 + x, label = "2x+x")
plt.plot(x, x**2 - x, label = "x^2 - x")
plt.legend(loc=4);
```

```python
plt.figure(figsize=(10,8), dpi=140)
plt.plot(x,x*2 + x, label = "2x+x")
plt.plot(x, x**2 - x, label = "x^2 - x")
plt.legend(loc=(1.1,1.5));
```

```python
plt.figure(figsize=(10,8), dpi=140)
plt.plot(x,x*2 + x, "r*-")
plt.plot(x, x**2 - x, "m.-")
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.plot(x,x+1,color="blue", linestyle="-", lw=5)
plt.plot(x,x+2,color="red", linestyle="-.", lw=2)
plt.plot(x,x+3,color="green", linestyle=":", lw=7)
```
